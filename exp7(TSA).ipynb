```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data = {
    'date': pd.date_range(start='2020-01-01', periods=365, freq='D'),
    'value': np.sin(np.linspace(0, 10, 365)) + np.random.normal(0, 0.1, 365)
}

df = pd.DataFrame(data)
df.set_index('date', inplace=True)
```

```python
result = seasonal_decompose(df['value'], model='additive', period=180)
```

```python
result.plot()
plt.show()
```

```python
trend = result.trend
seasonal = result.seasonal
residual = result.resid
```

```python
from statsmodels.tsa.seasonal import STL
```

```python
df = pd.DataFrame(data)
df.set_index('date', inplace=True)
```

```python
stl = STL(df['value'], seasonal=13)  # 13 is a window for seasonality
result_stl = stl.fit()
```

```python
result_stl.plot()
plt.show()
```

```python
trend_stl = result_stl.trend
seasonal_stl = result_stl.seasonal
residual_stl = result_stl.resid
```
