```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('shampoosales.csv')
```

```python
data=data.rename(columns={'Sales of shampoo over a three year period':'Sales'})
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Sales'])
plt.title('Shampoo Sales Over Time')
plt.xlabel('Date')
plt.xticks(rotation=45, fontsize=8)
plt.ylabel('Sales')
plt.show()
```

```python
monthly_data = data['Sales'].resample('M').mean() # Resample the 'Sales' column
plt.figure(figsize=(12, 6))
plt.plot(monthly_data)
plt.title('Monthly Shampoo Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data['Sales'].plot(kind='hist', bins=20, color='black', edgecolor='purple')
plt.title('Distribution of Shampoo Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data['Sales'].plot(kind='box')
plt.title('Box Plot of Shampoo Sales')
plt.ylabel('Sales')
plt.show()
```
